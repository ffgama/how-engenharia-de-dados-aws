## Segunda forma, descoberta de uma API interna

```python
import requests
import urllib.parse

search = 'coco'
base_url = 'https://marketing.condor.com.br/api/Produtos'

params = {"filter[include]": "ofertas",
          "filter[where][or][0][dsc_produto][like]": urllib.parse.quote_plus(f"%{search}%"),
          "filter[where][or][1][dsc_descricao][like]": urllib.parse.quote_plus(f"%{search}%")}

# Due to the way this API was done, some params
# are URL encoded, some not, so I will need to build query
# string by myself.

def params_to_string_url(params: dict):
    result = "?"
    first = True
    for k, v in params.items():
        if first:
            first = False
        else:
            result += '&'
        result += k
        result += '='
        result += v   

    return result

url = base_url + params_to_string_url(params)
```
